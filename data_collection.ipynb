```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from sqlalchemy import create_engine
```

```python
url = 'https://www.baseball-reference.com/teams/SFG/2022-batting.shtml'
params = {  }
```

```python
extended_request = requests.get(url, params=params)
extended_request
```

```python
print(extended_request.text)
```

```python
soup = BeautifulSoup(extended_request.text, 'html.parser')
print(soup.prettify())
```

```python
data = soup.find_all('table', attrs={'id':'team_batting'})
data
```

```python
rows = []
for element in data:
    row = element.find_all('tr')
    rows.extend(row)
rows
```

```python
stat = []
for element in rows:
    s = element.find('td',attrs={'data-stat':'G'})
    if s is not None:
      print(s.text)

```

```python
# loop through html
stats = {
    'player':[],
    'G':[],
    'PA':[],
    'AB':[],
    'R':[],
    'H':[],
    '2B':[],
    '3B':[],
    'HR':[],
    'RBI':[],
    'BB':[],
    'SO':[],
    'BA':[],
    'OBP':[],
    'SLG':[],
    'OPS':[],
    'OPS+':[],
    'TB':[],
    'HBP':[],
    'SH':[],
    'SF':[],
    'IBB':[]
}
```

```python
for element in rows:
    player = element.find('td',attrs={'data-stat':'player'})
    if player is not None:
      stats['player'].append(player.text)
    G = element.find('td',attrs={'data-stat':'G'})
    if G is not None:
      stats['G'].append(G.text)
    PA = element.find('td',attrs={'data-stat':'PA'})
    if PA is not None:
      stats['PA'].append(PA.text)
    AB = element.find('td',attrs={'data-stat':'AB'})
    if AB is not None:
      stats['AB'].append(AB.text)
    R = element.find('td',attrs={'data-stat':'R'})
    if R is not None:
      stats['R'].append(R.text)
    H = element.find('td',attrs={'data-stat':'H'})
    if H is not None:
      stats['H'].append(H.text)
    double = element.find('td',attrs={'data-stat':'2B'})
    if double is not None:
      stats['2B'].append(double.text)
    triple = element.find('td',attrs={'data-stat':'3B'})
    if triple is not None:
      stats['3B'].append(triple.text)
    HR = element.find('td',attrs={'data-stat':'HR'})
    if HR is not None:
      stats['HR'].append(HR.text)
    RBI = element.find('td',attrs={'data-stat':'RBI'})
    if RBI is not None:
      stats['RBI'].append(RBI.text)
    BB = element.find('td',attrs={'data-stat':'BB'})
    if BB is not None:
      stats['BB'].append(BB.text)
    SO = element.find('td',attrs={'data-stat':'SO'})
    if SO is not None:
      stats['SO'].append(SO.text)
    BA = element.find('td',attrs={'data-stat':'batting_avg'})
    if BA is not None:
      stats['BA'].append(BA.text)
    OBP = element.find('td',attrs={'data-stat':'onbase_perc'})
    if OBP is not None:
      stats['OBP'].append(OBP.text)
    SLG = element.find('td',attrs={'data-stat':'slugging_perc'})
    if SLG is not None:
      stats['SLG'].append(SLG.text)
    OPS = element.find('td',attrs={'data-stat':'onbase_plus_slugging'})
    if OPS is not None:
      stats['OPS'].append(OPS.text)
    OPSplus = element.find('td',attrs={'data-stat':'onbase_plus_slugging_plus'})
    if OPSplus is not None:
      stats['OPS+'].append(OPSplus.text)
    TB = element.find('td',attrs={'data-stat':'TB'})
    if TB is not None:
      stats['TB'].append(TB.text)
    HBP = element.find('td',attrs={'data-stat':'HBP'})
    if HBP is not None:
      stats['HBP'].append(HBP.text)
    SH = element.find('td',attrs={'data-stat':'SH'})
    if SH is not None:
      stats['SH'].append(SH.text)
    SF = element.find('td',attrs={'data-stat':'SF'})
    if SF is not None:
      stats['SF'].append(SF.text)
    IBB = element.find('td',attrs={'data-stat':'IBB'})
    if IBB is not None:
      stats['IBB'].append(IBB.text)
```

```python
stats
```

```python
df = pd.DataFrame(stats)
df.head()
```
